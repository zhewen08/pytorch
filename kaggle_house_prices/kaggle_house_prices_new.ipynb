```python
# Prepare data and features
#  #   Column                       Non-Null Count  Dtype',
# ---  ------                       --------------  -----',
#  0   Id                           79065 non-null  int64',
#  2   Sold Price                   47439 non-null  float64
#  4   Type                         79065 non-null  object',
#  5   Year built                   77123 non-null  float64
#  9   Lot                          56076 non-null  float64
#  10  Bedrooms                     74467 non-null  object',
#  11  Bathrooms                    73655 non-null  float64
#  12  Full bathrooms               66137 non-null  float64
#  13  Total interior livable area  75187 non-null  float64
#  18  Elementary School Score      70330 non-null  float64
#  21  Middle School Score          50786 non-null  float64
#  24  High School Score            71281 non-null  float64
#  32  Tax assessed value           72742 non-null  float64
#  33  Annual tax amount            71856 non-null  float64
#  34  Listed On                    79065 non-null  object',
#  35  Listed Price                 79065 non-null  float64
#  36  Last Sold On                 49520 non-null  object',
#  37  Last Sold Price              49520 non-null  float64
#  38  City                         79065 non-null  object',
#  39  Zip                          79065 non-null  int64',
#  40  State                        79065 non-null  object',

features = [
    'Year built',
    'Lot',
    'Bedrooms',
    'Bathrooms',
    'Full bathrooms',
    'Total interior livable area',
    'Elementary School Score',
    'Middle School Score',
    'High School Score',
    'Tax assessed value',
    'Annual tax amount',
    'Listed On',
    'Listed Price',
    'Last Sold On',
    'Last Sold Price',
    'City',
    'Zip',
    'State',
]

# Already excluded Id and label
all_features = all_data.loc[:, features]

# Process numeric columns
all_features['Bedrooms'] = pd.to_numeric(
    all_features['Bedrooms'], errors='coerce').astype('float64')
all_features['Listed On'] = pd.to_datetime(
    all_features['Listed On'], errors='coerce')
all_features['Last Sold On'] = pd.to_datetime(
    all_features['Last Sold On'], errors='coerce')
all_features['Zip'] = all_features['Zip'].astype('object')
# Datetime features needs further processing to be better numeric feature
# Skip datetime features for now
all_features = all_features.drop(['Listed On', 'Last Sold On'], axis=1)
# Simplify dataset
all_features = all_features.drop(['City', 'Zip', 'State'], axis=1)

numeric_features = all_features.select_dtypes('float64').columns
other_features = all_features.columns.difference(numeric_features)
print(numeric_features, other_features)

all_features[numeric_features] = all_features[numeric_features].fillna(
    all_features[numeric_features].mean()).apply(
    lambda x: (x - x.mean()) / x.std()).apply(
    lambda x: x.astype('float32'))
```

```python
# Process object columns - get_dummies which is computation intensive
# Type column has 500+ values
# Ele/mid/high schools have 5000+ values
all_features = pd.get_dummies(all_features, dummy_na=True)

# We use rest_data to train. Id is not a feature. Sold Price is label
ids = all_data['Id']
labels = all_data['Sold Price'].astype('float32')

all_features.info()
all_features.head(10)
```

```python
processed_data = pd.concat([ids, labels, all_features], axis=1)

groups = processed_data.groupby(
    processed_data['Sold Price'].notna())

labeled = groups.get_group(True)
unlabeled = groups.get_group(False)
```

```python
import torch
from torch.utils import data
import torch.nn as nn
import torch.nn.functional as F

# Define dataset and model
class HousePriceDataset(data.Dataset):
    def __init__(self, annotated_df, transform=None, 
                 target_transform=None):
        self.examples = annotated_df
        self.transform = transform
        self.target_transform = target_transform
        
    def __len__(self):
        return len(self.examples)
    
    def __getitem__(self, idx):
        example = self.examples.iloc[idx]
        # col1: Id, col2: Label (Sold Price), col rest: features
        X = torch.from_numpy(example[2:].values).float()
        y = torch.from_numpy(example[1:2].values).float()
        
        if self.transform:
            X = self.transform(X)
        if self.target_transform:
            y = self.target_transform(y)
            
        return X, y

class HousePriceNet(nn.Module):
    def __init__(self):
        super().__init__()
        num_inputs = 13
        num_hiddens1 = 6
        num_outputs = 1
        self.linear1 = nn.Linear(num_inputs, num_hiddens1)
        self.linear2 = nn.Linear(num_hiddens1, num_outputs)
        self.relu = nn.ReLU()
        self.verbose = False
        
    def forward(self, X):
        input = X
        if self.verbose:
            print(f'input: {input}')
        X = self.linear1(X)
        if self.verbose:
            print(f'linear1: {X}')
        X = self.relu(X)
        if self.verbose:
            print(f'relu: {X}')
        output = self.linear2(X)
        if self.verbose:
            print(f'linear2: {output}')
        return output
    
    def set_verbose(self, val):
        self.verbose = val
```

```python
# Train and validate

def train_epoch(net, train_iter, loss, updater):
    train_l = 0
    b = 0
    train_cnt = 0
    for X, y in train_iter:
        y_hat = net(X)
        l = loss(y_hat, y).sum()
        train_l += l
        updater.zero_grad()
        l.backward()
        updater.step()
        b += 1
        train_cnt += y.numel()
        if b % 9 == 0:
            print(f"batch {b}, train loss {train_l/train_cnt:8f}")
        
    return train_l / len(train_iter.dataset)

def validate(net, test_iter, loss):
    test_l = 0
    with torch.no_grad():
        for X, trues in test_iter:
            preds = net(X)
            test_l += loss(preds, trues).sum()

    return test_l / len(test_iter.dataset)

torch.manual_seed(42)

def y_transform(y):
    # log plus exp is very cpu intensive. it takes forever to process 40000 examples
    return y / 1000000

def rmse(y_hat, y):
    mse = nn.MSELoss()
    # print(torch.cat([y_hat, y, (y_hat - y) / y], axis=1))
    return mse((y_hat - y) / y, torch.zeros_like(y))

batch_size = 500
train_dataset = HousePriceDataset(t, target_transform=y_transform)
train_iter = data.DataLoader(train_dataset, batch_size, shuffle=True)
test_dataset = HousePriceDataset(v, target_transform=y_transform)
test_iter = data.DataLoader(test_dataset, batch_size, shuffle=False)

net = HousePriceNet()
loss = rmse
lr = 0.0005
updater = torch.optim.Adam(net.parameters(), lr)

epoch = 6
for e in range(epoch):
    train_loss = train_epoch(net, train_iter, loss, updater)
    test_loss = validate(net, test_iter, loss)
    print(f"epoch {e}, train loss {train_loss:8f}, test loss {test_loss:8f}")
```

```python
# Test and generate submission

sub_batch_size = 1000
sub_dataset = HousePriceDataset(unlabeled)
sub_iter = data.DataLoader(sub_dataset, sub_batch_size, shuffle=False)

def predict(net, sub_iter):
    net.eval()
    results = torch.Tensor([])
    for X, _ in sub_iter:
        preds = net(X) * 1000000
        results = torch.cat((results, preds), dim=0)
    return results

preds = predict(net, sub_iter)
```

```python
preds_df = pd.DataFrame(data=preds.detach().numpy(), columns=['Sold Price'])
preds_df.set_index(unlabeled['Id'].index, inplace=True)
submission = pd.concat([unlabeled['Id'], preds_df], axis=1)

len(submission), submission.head()
```
